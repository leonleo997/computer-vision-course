```python
%config Completer.use_jedi = False
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blankImage = np.zeros((512,512,3), dtype= np.int16)
```

```python
plt.imshow(blankImage)
```

```python
cv2.rectangle(blankImage, pt1=(384,10), pt2=(500,150),color=(0,255,0), thickness=10)
plt.imshow(blankImage)
```

```python
cv2.rectangle(blankImage,pt1=(200,200),pt2=(300,300), color=(0,0,255), thickness=5)
plt.imshow(blankImage)
```

```python
cv2.circle(blankImage, center=(100,100), radius=50, color=(255,0,0), thickness=8)
plt.imshow(blankImage)
```

```python
cv2.circle(blankImage, center=(100,100), radius=50, color=(255,0,0), thickness=8)
plt.imshow(blankImage)
```

```python
cv2.circle(blankImage, center=(400,400), radius=50, color=(255,0,0), thickness=3)
plt.imshow(blankImage)
```

```python
cv2.line(blankImage, pt1=(0,0), pt2=(512,512), color=(102,255,255), thickness=8)
plt.imshow(blankImage)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blankImage, text="Hello",org=(10,500),fontFace=font,fontScale=4,color=(255,255,255),
           thickness=3,lineType=cv2.LINE_AA)
plt.imshow(blankImage)
```

```python
blankImage = np.zeros((512,512,3), dtype= np.int32)
plt.imshow(blankImage)
```

```python
vertices=np.array([ [100,300], [200,200], [400,300], [200,400]], dtype=np.int32)
vertices
```

```python
vertices.shape
```

```python
pts.shape
```
